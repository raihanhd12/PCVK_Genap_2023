### 2. Tampilkan image dalam channel Red-Blue dan Green-Blue saja!


> **Red - Blue** = Pertama, kita membuat array kosong yang memiliki ukuran yang sama dengan gambar menggunakan` np.zeros_like(img_rgb)`. Setelah itu, kita mengatur nilai piksel pada channel yang diinginkan menjadi nol. Misalnya, pada gambar `img_redblue`, kita mengatur saluran warna Green pada semua piksel menjadi nol dengan cara `img_redblue[:,:,1] = 0`, sedangkan saluran warna Red dan Blue dibiarkan tetap menggunakan `img_redblue[:,:,0]` dan `img_redblue[:,:,2]`. 

> **Green-Blue** = Pertama, kita membuat array kosong yang memiliki ukuran yang sama dengan gambar menggunakan` np.zeros_like(img_rgb)`. Setelah itu, kita mengatur saluran warna Red pada semua piksel menjadi nol dengan `img_greenblue[:,:,0] = 0`, sedangkan saluran warna Green dan Blue dibiarkan tetap menggunakan `img_greenblue[:,:,1]` dan `img_greenblue[:,:,2]`. Terakhir, kita menampilkan kedua gambar yang sudah diproses dengan menggunakan fungsi `plt.imshow()` dan `plt.show()`.

> **Penjelasan** =  `[:,:,0]`, `[:,:,1]`, dan `[:,:,2]` adalah indeks untuk memilih channel pada gambar yang ditunjukkan dalam array NumPy. Pada saluran warna RGB atau BGR, saluran warna merah, hijau, dan biru masing-masing direpresentasikan oleh saluran ketiga dari dimensi gambar. Jadi, untuk memilih saluran warna tertentu, kita menggunakan indeks 0 untuk saluran warna merah, 1 untuk saluran warna hijau, dan 2 untuk saluran warna biru.

> **Kegunaan** `:` adalah untuk menunjukkan bahwa kita memilih semua baris dan kolom dari gambar, sedangkan angka pertama dalam indeks `:` menunjukkan indeks baris, dan angka kedua dalam indeks : menunjukkan indeks kolom. Sebagai contoh, `[:,:,0]` berarti kita memilih semua baris dan kolom dari gambar, dan saluran warna merah (indeks 0) pada gambar tersebut.

```python
# Inisiasi Variabel gambar
imgt2 = img

# Konversi channel BGR -> RGB
img_rgb = cv.cvtColor(imgt2, cv.COLOR_BGR2RGB)

# Tampilkan gambar dalam channel Red-Blue saja
img_redblue = np.zeros_like(img_rgb)
img_redblue[:,:,0] = img_rgb[:,:,0]
img_redblue[:,:,2] = img_rgb[:,:,2]
plt.imshow(img_redblue)
plt.show()

print("\n")

# Tampilkan gambar dalam channel Green-Blue saja
img_greenblue = np.zeros_like(img_rgb)
img_greenblue[:,:,1] = img_rgb[:,:,1]
img_greenblue[:,:,2] = img_rgb[:,:,2]
plt.imshow(img_greenblue)
plt.show()
```

### 5. Buat 5 kotak berbagai ukuran dan warna yang berbeda dalam satu image.disarankan menggunakan bilangan acak/random!

> **Penjelasan** : Untuk membuat proses pembuatan lima kotak acak dilakukan dalam loop for dengan lima iterasi. Pada setiap iterasi, ukuran kotak, warna kotak, dan posisi kiri atas kotak akan diacak menggunakan fungsi `np.random.randint`. `np.random.randint` digunakan untuk menghasilkan bilangan bulat acak dalam range tertentu.

> **Pertama** : Pada contoh dibawah ukuran kotak diacak antara 10 dan 100 piksel.

> **Kedua** : Warna kotak diacak dalam range 0 sampai 255 untuk setiap channel warna (RGB).

> **Ketiga** : Posisi kiri atas kotak juga diacak dalam range antara (0,0) dan (299 - ukuran kotak, 299 - ukuran kotak) untuk memastikan kotak tidak melebihi ukuran image.

> **Keempat** : Setelah ukuran, warna, dan posisi kotak dihasilkan secara acak, Kode `cv.rectangle(img_rgb, (x, y), (x + size, y + size), tuple(color.tolist()), thickness=3)` digunakan untuk menggambar sebuah persegi panjang berwarna pada image dengan menggunakan OpenCV (cv).

> * `img_rgb` adalah variabel yang berisi image yang akan digambar, `(x, y) ` adalah titik kiri atas dari persegi panjang, `(x + size, y + size)` adalah titik kanan bawah dari persegi panjang, dengan size adalah ukuran persegi panjang. `tuple(color.tolist())` adalah sebuah tuple berisi tiga nilai integer yang merepresentasikan warna yang digunakan untuk menggambar persegi panjang tersebut dan` color.tolist()` digunakan untuk mengonversi nilai numpy array `color` menjadi list dan kemudian mengonversi list tersebut menjadi tuple, `thickness=3` adalah ketebalan garis yang digunakan untuk menggambar persegi panjang.

> Jadi, keseluruhan program menghasilkan image dengan lima kotak acak berbagai ukuran dan warna pada posisi acak pula.

```python
# Inisiasi Variabel gambar
imgt5 = img

# Konversi channel BGR -> RGB
img_rgb = cv.cvtColor(imgt5, cv.COLOR_BGR2RGB)

# Membuat 5 kotak berbagai ukuran, posisi, dan warna dengan ketebalan garis 3
for i in range(5):
    # Ukuran kotak acak antara 10 dan 100
    size = np.random.randint(10, 100)
    
    # Warna kotak acak
    color = np.random.randint(0, 256, size=(3,))
    
    # Posisi kiri atas kotak acak antara (0, 0) dan (299 - size, 299 - size)
    x = np.random.randint(0, 300 - size)
    y = np.random.randint(0, 300 - size)
    
    # Menggambar kotak pada image dengan ketebalan garis 3
    cv.rectangle(img_rgb, (x, y), (x + size, y + size), tuple(color.tolist()), thickness=3)

# Menampilkan image
plt.imshow(img_rgb)
plt.show()
```

1. cv.FONT_HERSHEY_SIMPLEX : jenis font yang digunakan, dalam hal ini menggunakan font "Hershey Simplex".

2. text : teks yang akan ditambahkan pada gambar.
